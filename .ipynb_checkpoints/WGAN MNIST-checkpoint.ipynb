# Discriminator NN
Similar to GAN,but the output layer and the weight constraint.  
![Wasserstein Distance](https://1.bp.blogspot.com/-0wsj0c3Q5Nw/XWUsbJ-fxqI/AAAAAAAAEiI/JjWSkU9Dh_sJrYhkWISrWpbGVx7mXliYACLcBGAs/s640/%25E6%2593%25B7%25E5%258F%2596.PNG) 
* loss function並非cross entropy(or js divergence)，為wasserstein distance
* output layer不經過sigmoid，直接輸出原始值  

2 conv + 1 fully  

```python
def discriminator(input_shape=(28,28,1)):
    init=keras.initializers.RandomNormal(stddev=0.02)
    clip=weightClipping(0.01)
    
    model=keras.models.Sequential([
        #to(14*14*64)
        keras.layers.Conv2D(64,(4,4),strides=(2,2),padding='same',kernel_initializer=init,kernel_constraint=clip,input_shape=input_shape),
        keras.layers.BatchNormalization(),
        keras.layers.LeakyReLU(0.2),
        #to(7*7*128)
        keras.layers.Conv2D(64,(4,4),strides=(2,2),padding='same',kernel_initializer=init,kernel_constraint=clip),
        keras.layers.BatchNormalization(),
        keras.layers.LeakyReLU(0.2),
        
        keras.layers.Flatten(),
        keras.layers.Dense(1,activation='linear')
    ])
    model.compile(loss=wassersteinDistance,optimizer=keras.optimizers.RMSprop(lr=0.00005))
    return model
```

```python
class weightClipping(keras.constraints.Constraint):
    def __init__(self,c):
        self.clip_value=c
    def __call__(self,weights):
        return keras.backend.clip(weights,-self.clip_value,self.clip_value)
```

```python
def wassersteinDistance(y_true,y_pred):
    return keras.backend.mean(y_true*y_pred)
```

# Generator NN
similar to GAN  

```python
def generator(input_dim):
    init=keras.initializers.random_normal(stddev=0.02)
    
    model=keras.models.Sequential([
        keras.layers.Dense(7*7*128,kernel_initializer=init,input_dim=input_dim),
        keras.layers.LeakyReLU(0.2),
        keras.layers.Reshape((7,7,128)),
        
        #to(14*14*128)
        keras.layers.Conv2DTranspose(128,(4,4),strides=(2,2),padding='same',kernel_initializer=init),
        keras.layers.BatchNormalization(),
        keras.layers.LeakyReLU(0.2),
        #to(28*28*128)
        keras.layers.Conv2DTranspose(128,(4,4),strides=(2,2),padding='same',kernel_initializer=init),
        keras.layers.BatchNormalization(),
        keras.layers.LeakyReLU(0.2),
        
        keras.layers.Conv2D(1,(7,7),activation='tanh',padding='same',kernel_initializer=init)
    ])
    return model
```

# Composite GAN NN

```python
def gan(g_model,d_model):
    d_model.trainable=False
    model=keras.models.Sequential([g_model,d_model])
    model.compile(loss=wassersteinDistance,optimizer=keras.optimizers.RMSprop(lr=0.00005))
    return model
```

# Image_Generator(for Discriminator)
* real label -1
* fake label +1

__but why???__  
我先用相反的: real=1,fake=-1  
https://github.com/keras-team/keras-contrib/issues/280  

但好像沒差！  

```python
def real_image_generator(dataset,num):
    randi=np.random.randint(0,len(dataset),num)
    X=dataset[randi]
    y=np.ones((num,1))
    return X,y
```

```python
def fake_image_generator(g_model,num,latent_dim):
    z=np.random.randn(num,latent_dim)
    X=g_model.predict(z)
    y=-np.ones((num,1))
    return X,y
```

# Train
By wasserstein distance objective function，loss_real越大越好，loss_fake越小越好  

```python
def fit(g_model,d_model,dataset,latent_dim,epochs=100,batch_size=64):
    gan_model=gan(g_model,d_model)
    batchs=len(dataset)//batch_size
    loss_real_his=[];loss_fake_his=[];loss_generator_his=[]
    for epoch in range(1,epochs+1):
        for batch in range(1,batchs+1):
            #train discriminator
            r=f=0
            for t in range(5):
                X,y=real_image_generator(dataset,batch_size//2)
                loss_real=d_model.train_on_batch(X,y)
                r+=loss_real
                X,y=fake_image_generator(g_model,batch_size//2,latent_dim)
                loss_fake=d_model.train_on_batch(X,y)
                f+=loss_fake
            loss_real_his.append(r/5)
            loss_fake_his.append(f/5)
            
            #train generator
            z=np.random.randn(batch_size,latent_dim)
            y=np.ones((batch_size,1))
            loss_generator=gan_model.train_on_batch(z,y)
            loss_generator_his.append(loss_generator)
            
            #print loss
            print('epoch{} {}/{} loss_real:{:.2f} loss_fake:{:.2f} loss_generator:{:.2f}'.format(epoch,batch,batchs,loss_real_his[-1],loss_fake_his[-1],loss_generator_his[-1]))
        if epoch%5==0:
            summarize(epoch,g_model,latent_dim)
    return loss_real_his,loss_fake_his,loss_generator_his
```

```python
def summarize(epoch,g_model,latent_dim):
    #save generate image
    z=np.random.randn(25,latent_dim)
    X=g_model.predict(z)
    #from [-1,1] to [0,1]
    X=(X+1)/2 
    plt.figure()
    for i in range(25):
        plt.subplot(5,5,i+1)
        plt.imshow(X[i,:,:,0],cmap='gray')
    plt.savefig('WGAN MNIST\\epoch{}.png'.format(epoch))
    plt.close()
    #save model
    g_model.save('WGAN MNIST\\generator epoch{}.h5'.format(epoch))
```

# Start WGAN

### load data

```python
(train_X,train_y),(_,_)=load_data()
train_X=train_X[train_y==5]
train_X=np.expand_dims(train_X,axis=-1)
#from [0,255] to [-1,1]
train_X=(train_X-127.5)/127.5
```

### define latent dimensioin

```python
latent_dim=50
```

### train

```python
g_model=generator(latent_dim)
d_model=discriminator(train_X[0].shape)
loss_real_his,loss_fake_his,loss_generator_his=fit(g_model,d_model,train_X,latent_dim,epochs=100)
```

### plot loss value
_A well-performing WGAN should show this line trending down as the image quality of the generated model is increased._  
* loss_real越大越好，loss_fake越小越好  
* loss_generator越小越好  

```python
plt.figure()
plt.plot(loss_real_his,label='loss_real')
plt.plot(loss_fake_his,label='loss_fake')
plt.plot(loss_generator_his,label='loss_generator')
plt.legend()
plt.savefig('WGAN MNIST\\plot loss.png')
```
